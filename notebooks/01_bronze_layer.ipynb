# Bronze Layer

```python
# Importing all the required functions
from pyspark.sql.functions import current_timestamp
```

```python
# Reading the raw data without any changes
bronze_df = spark.read\
            .format("csv")\
            .option('header', "true")\
            .option("inferSchema", "true")\
            .load("/Volumes/workspace/default/aqi/aqi_data.csv")            
```

```python
# Adding a new column for the ingestion time
bronze_df = bronze_df.withColumn("ingestion_timestamp", current_timestamp())
```

```python
# Writing the data to the table as Delta format
bronze_df.write\
  .format("delta")\
  .mode("overwrite")\
  .saveAsTable("aqi_bronze")
```
